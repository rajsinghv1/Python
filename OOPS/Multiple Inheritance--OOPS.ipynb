### OOPS-Multiple Inheritance in Python

```python
class A:
    def method1(self):
        print("A class method is called")
```

```python
class B(A):
    def method1(self):
        print("B class method is called")
    def method2(self):
        print("B class method is called")
```

```python
class C(A):
    def method1(self):
        print("C class method is called")
```

```python
class D(B,C):
    def method1(self):
        print("D class method is called")
```

```python
d=D()
```

```python
d.method2()
```

```python
B.method1(d)
```

```python
A.method1(d)
```
