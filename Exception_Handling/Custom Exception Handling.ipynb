### Custom Exception

```python
class Error(Exception):
    pass

class dobException(Error):
    pass

```

```python
year=int(input("Enter the year of Birth"))
age=2021-year
try:
    if age<=30 & age>20:
        print("The age is valid. You can apply for the exams")
    else:
        raise dobException

except dobException:
    print("The year age is not valid. You cannot apply for the exams")
```
